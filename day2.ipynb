# Basic built-in collections -- summary

Cf. also the relevant sections of your [cheat sheet](https://perso.limsi.fr/pointal/_media/python:cours:mementopython3-english.pdf).

Collections are also called **containers**. Ordered collections are also called **sequences**.

- `str`: an immutable ordered collection of characters; values can repeat
- `list`: a mutable ordered collection of any type of object; values can repeat
- `tuple`: an immutable ordered collection of any type of object; values can repeat
- `set`: a mutable unordered collection of (almost) any type of object; values are unique, duplicates are discarded when you try to insert them
- `dict`: a mutable unordered collection mapping keys to values
    - keys can be (almost) any type of object and they are unique (kind of like numerical indices are unique in a list)
    - values can be any type of object and they can repeat
    
A `set` is kind of like a `dict` where the keys aren't mapped to any useful values (plus it has a lot of specific methods for performing useful set operations, like unions, intersections etc.).

An important property of sets and dicts is that checking whether they contain a given key is **really** fast, even if the collection is large. By comparison, with ordered collections (`str`, `list`, `tuple`), checking whether they contain a given item amounts to inspecting the items one by one, which can take a long time. If you want, you can picture this as the difference between searching for a word in a normal language dictionary (`dict`, `set`) versus in a dictionary that has not been alphabetically sorted (`str`, `list`, `tuple`).

````python
wordcloud(
    word  # 3.
    for word in text6  # 1.
    if word.isalpha() and not word.isupper()  # 2. (optional)
)
````

**How to read the following piece of code:** For each word in text6 (1.), if that word is alphabetic and not uppercase (2.), take the lowercase representation of that word (3.) and put it into a wordcloud.

Or, in the same order that it's written in the code: Take the lowercase variant of the word (3.) for each word in text6 (1.), but only if that word is alphabetic and not uppercase (2.).

Again, `word` is a name that we pick ourselves. It could be `x` or `foo` or `whatever`.

````python
wordcloud(
    word.lower()  # 3.
    for word in text6  # 1.
    if word.isalpha() and not word.isupper()  # 2. (optional)
)
````

There's special syntax for creating new `list`s, `dict`s and `set`s using generator expressions. These are called `list` / `dict` / `set` comprehensions.

A **list comprehension** is basically a generator expression wrapped in square brackets:

````python
orig_list = ["Call", "me", "Ishmael", ".",
             "Call", "him", "Moby-Dick", "."]
````

````python
[x for x in orig_list if x.istitle()]
````

````python
[1, 2, 3]
````

A **set comprehension** is a generator expression wrapped in curly brackets:

````python
{x for x in orig_list if x.istitle()}
````

A **dict comprehension** is *kind of* like a generator expression wrapped in curly brackets, because you need to use the `key: value` syntax:

````python
{x: len(x) for x in orig_list if x.istitle()}
````

````python
{x: len(x)
 for x in orig_list
 if x.istitle()}
````

A **character set** is kind of like a dictionary: it's a big table that maps characters to numbers. Let's create our own little toy character set and call it e.g. `joe`:

````python
charset_joe = {
    "a": 1,
    "b": 2,
    "c": 3,
    "d": 4
}
````

An **encoding** is kind of like a function which takes a character and returns a series of 1's and 0's -- bits -- which can be stored in the computer's memory to represent the character. Let's create an encoding for our toy character set, `joe`, and call it e.g. `eve`:

````python
def eve(char):
    num = charset_joe[char]
    if num == 1:
        return "1"
    elif num == 2:
        return "11"
    elif num == 3:
        return "111"
    elif num == 4:
        return "1111"
````

````python
eve("b")
````

````python
[eve(char) for char in "ab"]
````

````python
"111"
````

Uh-oh, the `eve` encoding is no good -- trying to encode the string `"ab"`, it would store `111` in the computer's memory, but if we tried to read that back, it would not be clear whether it should be interpreted as `"aaa"`, `"ab"`, `"ba"` or `"c"`.

````python
def eve2(char):
    num = charset_joe[char]
    if num == 1:
        return "10"
    elif num == 2:
        return "110"
    elif num == 3:
        return "1110"
    elif num == 4:
        return "11110"
````

````python
[eve2(char) for char in "ab"]
````

Fixed-width encoding

````python
def eve3(char):
    num = charset_joe[char]
    if num == 1:
        return "00"
    elif num == 2:
        return "01"
    elif num == 3:
        return "10"
    elif num == 4:
        return "11"
````

````python
[eve3(char) for char in "ab"]
````

````python
def eve4(char):
    num = charset_joe[char] - 1
    return bin(num)
````

````python
[eve4(char) for char in "ab"]
````

````python
num = 1
# everything after the : is a format specifier
f"{num:02b}"
````

````python
def eve5(char):
    num = charset_joe[char] - 1
    return f"{num:02b}"
````

````python
[eve5(char) for char in "ab"]
````

## Unicode: a universal character set

Over 1 million characters (and counting)!

The Unicode **character set** can be encoded (= transformed into a sequence of 1's and 0's that can be stored in a computer's memory) using several different **encodings**:

- (more or less) fixed-width ones, like UTF-16 or UTF-32
- a really clever variable-width one, UTF-8, which is ASCII-compatible and is becoming ubiquitous (use it!)

````python
bin(65)
````

````python
def print_utf8_bytes(char):
    """Prints binary representation of character as encoded by UTF-8.
    
    """
    # encode the string as UTF-8 and iterate over the bytes;
    # iterating over a sequence of bytes yields integers in the
    # range [0; 256); the formatting directive "{:08b}" does two
    # things:
    #   - "b" prints the integer in its binary representation
    #   - "08" left-pads the binary representation with 0's to a total
    #     width of 8, which is the width of a byte
    binary_bytes = [f"{byte:08b}" for byte in char.encode("utf8")]
    print(f"{char!r} encoded in UTF-8 is: {binary_bytes}")

print_utf8_bytes("A")   # the representations...
print_utf8_bytes("č")   # ... keep...
print_utf8_bytes("字")  # ... getting longer.
````

A plain-text file has no information about which encoding it's in, you either have to know, or try several plausible ones until the result is what you'd expect. This looks right...

````python
with open("/home/lukes/edu/v4py/text.utf8", encoding="utf-8") as file:
    for line in file:
        print(line, end="")
````

... and this doesn't:

````python
with open("/home/lukes/edu/v4py/text.utf8", encoding="latin2") as file:
    for line in file:
        print(line, end="")
````

When you don't know the encoding, always try UTF-8 first, because it has some built-in checks that mean that not any old sequence of bits can be interpreted as UTF-8 data. This means that if the file can be read in as UTF-8 without errors, there's a good chance it actually *is* UTF-8. Conversely, if errors are raised (like below), you know for sure that it's a different encoding:

````python
with open("/home/lukes/edu/v4py/text.latin2", encoding="utf-8") as file:
    for line in file:
        print(line, end="")
````
